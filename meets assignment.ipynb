```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('meets.csv')
```

```python
print(df.head())
```

```python
print(df.describe())
```

```python
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
sns.pairplot(df)
plt.show()
```

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
sns.histplot(data=df, x='MeetState', kde=True)
plt.title('Distribution of MeetPath')
plt.xlabel('MeetState')
plt.ylabel('MeetTown')
plt.show()
```

```python
sns.countplot(data=df, y='Federation')
plt.title('Number of Federation')
plt.xlabel('Count')
plt.ylabel('MeetState')
plt.xticks(rotation=45)
plt.show()
```
